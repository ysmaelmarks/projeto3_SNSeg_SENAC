```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dados = pd.read_csv('https://raw.githubusercontent.com/ysmaelmarks/projeto3_SNSeg_SENAC-/main/dados.csv')
df = pd.DataFrame(dados)
```

# Códigos e cálculos

Total de Horas Trabalhadas

```python
horasTrabalhadas = df['Horas Trabalhadas'].sum()
print(f'Horas trabalhadas: {horasTrabalhadas}')
```

Média Diária de Horas Trabalhadas

```python
mediaHorasTrabalhadas = df['Horas Trabalhadas'].mean()
print(f'Média diária de: {mediaHorasTrabalhadas:.2f} horas')
```

Total de Bugs Corrigidos

```python
bugsCorrigidos = df['Bugs Corrigidos'].sum()
print(f'Bugs corrigidos: {bugsCorrigidos}')
```

Média Diária de Bugs Corrigidos

```python
print(f'Bugs corrigidos: {bugsCorrigidos/7:.1f}')
```

Total de Tarefas Concluídas

```python
tarefasConcluidas = df['Tarefas Concluídas'].sum()
print(f'Tarefas concluídas: {tarefasConcluidas}')
```

Média Diária de Tarefas Concluídas


```python
mediaTarefasConcluidas = df['Tarefas Concluídas'].mean()
print(f'Média diária de: {mediaTarefasConcluidas:.1f} tarefas concluídas')
```

Produtividade Diária

```python
produtividade = tarefasConcluidas / horasTrabalhadas
print(f'Produtividade diária de: {produtividade:.1f} tarefas concluídas por hora')
```

# Insights

## Gráfico de Horas Trabalhadas

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Horas Trabalhadas'], marker='o')
plt.title('Horas Trabalhadas')
plt.xlabel('Índice')
plt.ylabel('Horas Trabalhadas')
plt.grid(True)
plt.show()
```

### Consistência
Podemos verificar se há consistência no número de horas trabalhadas por dia. Picos ou vales podem indicar períodos de maior ou menor esforço.


### Padrões
Podemos identificar se existem padrões de trabalho, como horários de pico durante certos dias da semana.

## Gráfico de Bugs Corrigidos

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Bugs Corrigidos'], marker='x', color='r')
plt.title('Bugs Corrigidos')
plt.xlabel('Índice')
plt.ylabel('Bugs Corrigidos')
plt.grid(True)
plt.show()
```

### Eficiência na Correção de Bugs
Podemos avaliar a eficiência da equipe em corrigir bugs. Um aumento constante pode indicar melhoria nas habilidades de resolução de problemas.

### Períodos Problemáticos
Podemos identificar períodos em que muitos bugs foram corrigidos, o que pode indicar problemas anteriores significativos no código

## Gráfico de Tarefas Concluídas

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Tarefas Concluídas'], marker='s', color='g')
plt.title('Tarefas Concluídas')
plt.xlabel('Índice')
plt.ylabel('Tarefas Concluídas')
plt.grid(True)
plt.show()
```

### Produtividade Diária
Podemos avaliar a produtividade diária da equipe. Um número consistente de tarefas concluídas sugere boa gestão de tempo e foco.

### Variáveis Externas
Podemos comparar a produtividade com eventos externos conhecidos (como reuniões, lançamentos, etc.) para entender o impacto desses eventos na produtividade.

## Gráfico de Produtividade Diária

```python
df['Produtividade'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Produtividade'], marker='d', color='m')
plt.title('Produtividade Diária')
plt.xlabel('Índice')
plt.ylabel('Tarefas Concluídas por Hora')
plt.grid(True)
plt.show()
```

### Eficiência do Trabalho
Podemos medir a eficiência do trabalho ao longo do tempo. A produtividade crescente indica melhoria nos processos ou habilidades da equipe.

### Identificação de Ineficiências
Podemos observar quedas na produtividade, que podem sinalizar problemas como cansaço, falta de recursos ou necessidade de treinamento adicional.
